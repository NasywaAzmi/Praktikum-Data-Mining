##**Exploratory Data Analysis**

```python
country = df[['Country/Territory', 'Malaria']]
country = country.groupby('Country/Territory').median()
country = country.rename_axis('Country/Territory').reset_index()
country = country.sort_values(by=['Malaria'], ascending=False).head(10)
country
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(country['Country/Territory'], country['Malaria'])

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)

# Add x, y gridlines

# Show top values
ax.invert_yaxis()

plt.show()
```
